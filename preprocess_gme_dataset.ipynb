```python
# renaming columns
new_column_names = {
    'Date': 'date',
    'Open': 'open_price',
    'High': 'high_price',
    'Low': 'low_price',
    'Close': 'close_price',
    'Adj Close': 'adjclose_price',
    'Volume': 'volume'
}

gme_df.rename(columns=new_column_names, inplace=True)
```

```python
column_names = gme_df.columns.tolist()
print(column_names)
```

```python
gme_df['previous_close'] = gme_df['close_price'].shift(1)
```

```python
column_names = gme_df.columns.tolist()
print(column_names)
```

```python
gme_df.describe
```

```python
gme_df['price_movement'] = ((gme_df['close_price'] - gme_df['previous_close']) / gme_df['previous_close']) * 100
```

```python
gme_df.describe
```

```python
gme_df['price_movement_tomorrow'] = gme_df['price_movement'].shift(1)
```

```python
gme_df = gme_df[::-1]
```
